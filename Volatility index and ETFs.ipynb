# Volatility index and associated ETFs explained

In this guide, we'll break down the Volatility Index (VIX) and how to make sense of it. We'll also look at ETFs related to the VIX, like VXX and UVXY, and how they can be used in trading strategies.

## VIX

### What is the VIX and why does it matter?
The Volatility Index (VIX), often referred to as the "fear index" was launched in 1993. It represents the market's expectation of volatility (known as implied volatility) over the coming 30 days. 

When the VIX is low, volatility is low. When the VIX is high, it typically signals a time when the market is gripped by fear. The VIX generally rises when stocks fall and declines when stocks rise. This indicates a strong negative correlation between the VIX and stock market returns.

Investors use the VIX to measure the level of risk, fear, or stress in the market when making investment decisions. Traders use the VIX to gauge market sentiment, helping them assess potential risks and decide whether to hedge or make directional trades.

For the past several years, if the VIX was trading below 20, the market was considered stable, while levels of 30 or more indicated high volatility. Still, remember that trading volatility is not the same as trading a market downturn; the market can decline even if volatility remains low.

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Download historical data for S&P 500 and VIX
sp500_data = yf.download("^GSPC", start="2020-01-01")
vix_data = yf.download("^VIX", start="2020-01-01")

# Align data based on dates
data = pd.DataFrame({
    'S&P 500': sp500_data['Close'],
    'VIX': vix_data['Close']
}).dropna()

# Plot S&P 500 and VIX on the same graph with dual y-axes
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot S&P 500 on primary y-axis (left)
ax1.plot(data.index, data['S&P 500'], label='S&P 500', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('S&P 500 Index Level', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

# Create secondary y-axis for VIX (right)
ax2 = ax1.twinx()
ax2.plot(data.index, data['VIX'], label='VIX', color='red')
ax2.set_ylabel('VIX Index Level', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title and legends
fig.suptitle('S&P 500 and VIX Index')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

plt.show()

```

### How is the VIX is calculated?
The VIX represents the annualized implied volatility of a hypothetical option on the S&P500 stock index with 30 days to expiration, based on the prices of near-term S&P500 options traded on the Chicago Board Options Exchange (CBOE). The calculation includes options that expire on the third Friday of each month, as well as weekly CBOE SPX options that expire every Friday.

The VIX is calculated using the following formula:

$$
\text{VIX} = 100 \times \sqrt{\frac{2}{T} \sum_{i} \frac{\Delta K_i}{K_i^2} e^{RT} Q(K_i) - \frac{1}{T} \left( \frac{F}{K_0} - 1 \right)^2 }
$$

Where:
- $T$ = Time to expiration
- $F$ = Forward index level derived from index option prices
- $K_0$ = First strike below the forward index level $F$
- $\Delta K_i$ = The interval between strike prices
- $K_i$ = Strike price of the $i$ th out-of-the-money option (put if $K_i < F$, call if $K_i > F$)
- $R$ = Risk-free interest rate
- $Q(K_i)$ = Midpoint of the bid-ask spread for each option with strike $K_i$

By combining the weighted prices of multiple S&P 500 put and call options across a wide range of strike prices, the VIX offers insight into the prices traders are willing to pay to buy or sell the S&P 500. These final values estimate the future volatility of the S&P 500. For example, if the VIX index is 22, it indicates that a hypothetical S&P 500 option with 30 days to expiration has an annualized implied volatility of 22%.

The options included in the calculation are typically at-the-money, reflecting the general market perception of which strike prices are most likely to be hit before expiry. This provides insight into the broader market sentiment regarding the direction of the market.

In summary, the VIX measures how much people are willing to pay to buy or sell the S&P 500, with the more they are willing to pay indicating more uncertainty.

### How Traders Use the VIX
Although the VIX is commonly discussed on a spot basis, no ETF or ETN directly represents spot VIX volatility. Instead, there are ETFs/ETNs that track collections of VIX futures, which only roughly approximate the VIXs performance.

Here are examples of trades and strategies using the VIX:

- <b>High VIX (>30)</b> - Buy put options or inverse ETFs to profit from further declines or hedge against increased market volatility.
- <b>Rising VIX</b> - Buy volatility ETFs to profit from the rise. Increase cash holdings to reduce exposure to risky assets. Buy put options on existing positions to protect against downside risk.
- <b>Low VIX (<20)</b> - Buy call options or long ETFs to profit from market gains. Increase exposure to stocks. Sell covered calls to generate income through premiums. 
- <b>Falling VIX</b> - Take long positions in equities. Short volatility ETFs to profit from the decline in volatility. 
- <b>VIX spikes</b> - Take advantage of short-term price movements with quick trades. Use options strategies (like straddles or strangles) to profit from large price swings in either direction.


## Futures
Before exploring Volatility ETFs it is important to understand the concept of futures. 
A futures contract is an agreement to buy or sell an asset at a specific price on a future date. Futures are used across different markets (commodities, stocks, volatility, etc.), and the idea is to lock in a price today for something that will be delivered or settled in the future. E.g. If you think the price of oil is going to rise, you might buy a futures contract today to lock in a lower price for future delivery.

The main difference between a futures contract and an options contract is that with a futures contract, you are obligated to fulfill the terms of the contract, whereas with an options contract, you have the right but not the obligation to execute the contract.

### How does this apply to VIX?

The VIX measures current volatility, but you cant directly trade the VIX itself. Instead, you trade VIX futures, which represent market expectations for volatility in the future.

So, if youre buying VIX futures, youre essentially betting on what volatility will be in, say, one month or two months from now.

### Contango and Backwardation

Contango and backwardation are concepts that describe the relationship between the futures price and the expected spot price (the price in the present market) as the contract nears expiration.

Contango happens when the futures price is higher than the expected spot price at contract maturity. In other words, future contracts are more expensive than what the asset is expected to be worth in the present market when they mature.

Backwardation occurs when futures prices are lower than the expected spot price. In this case, near-term futures are cheaper than what the market expects the spot price to be at contract expiration.


## VXX
VXX is an exchange-traded note (ETN). An ETN is similar to an ETF, but instead of holding a basket of stocks, bonds, or commodities, the ETN is a debt instrument with a maturity date.

VXX does not track the VIX itself (spot VIX); instead, it tracks the S&P 500 VIX Short-Term Futures Total Return Index, which is based on VIX futures.

### How VXX Tracks the S&P 500 VIX Short-Term Futures Total Return Index
The S&P 500 VIX Short-Term Futures Total Return Index is designed to provide exposure to equity market volatility through the VIX Index futures.

The index offers exposure to a daily rolling long position in the first and second month VIX futures contracts, reflecting market participants views on the future direction of the VIX Index at the time those futures contracts expire.

Since VXX must roll its futures contracts to rebalance the fund to the later contract, the fund manager is forced to sell the futures contracts that are closest to their expiration dates and buy the next dated contracts, which is a process called rolling. Since longer-dated futures contracts are often at higher levels than shorter-dated ones (during normal market conditions), the rolling activity can result in losses (as the ETN is forced to sell the lower-valued contracts and buy the higher-priced contracts).

Below is a graph comparing VXX and VIX:

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Download historical data for VXX and VIX
vxx_data = yf.download("VXX", start="2020-01-01")
vix_data = yf.download("^VIX", start="2020-01-01")

# Align data based on dates
data = pd.DataFrame({
    'VXX': vxx_data['Close'],
    'VIX': vix_data['Close']
}).dropna()

# Plot the data with dual y-axes
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot VXX on the left y-axis
ax1.plot(data.index, data['VXX'], label='VXX', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('VXX Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

# Create a secondary y-axis for VIX on the right side
ax2 = ax1.twinx()
ax2.plot(data.index, data['VIX'], label='VIX', color='red')
ax2.set_ylabel('VIX Index Level', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title and legends
fig.suptitle('VXX vs. VIX Index')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

plt.show()
```

This graph shows how VXX decays due to the rolling costs, highlighting why many traders often short VXX.

### Contango and Backwardation
Since VXX rolls futures contracts monthly, it buys new (more expensive) contracts as old ones expire. This process creates a loss over time as they constantly sell lower-priced contracts and buy higher-priced ones. This is called <b>contango</b> and is the main reason VXX decays over time. 

<b>Backwardation</b> usually happens during periods of extreme volatility or stress.For example, if theres a sudden spike in volatility, near-term futures might be priced lower than current spot VIX levels, reflecting expectations that the market will soon stabilize. In this scenario, VXX benefits because the fund is buying cheaper contracts and selling them at a higher price as they approach maturity. Although backwardation is less common, it can lead to an increase in the ETNs value when it occurs.

The example below simulates both contango (where second-month futures are more expensive) and backwardation (where short-term futures spike higher).



```python
# Example to demonstrate contango and backwardation
import numpy as np

# Generate example data for VIX futures prices
dates = pd.date_range(start="2023-01-01", periods=100)
first_month_futures = np.linspace(20, 30, len(dates))  # Gradual increase (contango)
second_month_futures = first_month_futures + 2  # Higher prices (contango)

# Simulate a backwardation scenario at the end
second_month_futures[-20:] = first_month_futures[-20:] - 3

# Plot the contango and backwardation example
plt.figure(figsize=(14, 7))
plt.plot(dates, first_month_futures, label='First Month VIX Futures', color='green')
plt.plot(dates, second_month_futures, label='Second Month VIX Futures', color='red')
plt.axvline(dates[-20], color='black', linestyle='--', label='Backwardation Starts')
plt.title('Contango and Backwardation Example')
plt.xlabel('Date')
plt.ylabel('VIX Futures Price')
plt.legend()
plt.grid(True)
plt.show()
```

In simple terms:
- In Contango - If the market is calm, VIX futures are usually in contango. VXX will decay over time because they must constantly roll into more expensive contracts, leading to gradual losses.

- In Backwardation - If theres a sudden market shock, VIX futures can enter backwardation. In this scenario, VXX might gain value because they are selling contracts at higher prices and rolling into cheaper ones.

### Common Trading Strategies Using VXX
VXX is often used by traders to bet on rising volatility, hedge against market downturns, or speculate on short-term moves in the VIX. Some common strategies include:

- Going Long: Traders may buy VXX to profit from expected spikes in volatility, typically during periods of market uncertainty or fear.
- Shorting VXX: Given VXXs natural tendency to decay due to contango, some traders short VXX, betting on the long-term downward trend.
- Hedging: Investors might use VXX as a hedge against a broader portfolio during times of expected market turbulence.

In summary, VXX is more a bet on the expected future volatility rather than the immediate level of volatility. As a result:

- During periods of market stress, when the VIX spikes, VXX tends to risebut not as sharply as the VIX itself.
- Over time, VXX tends to decay due to contango, as it constantly buys higher-priced longer-term contracts and sells lower-priced near-term contracts.

## UVXY
UVXY is a leveraged ETF that aims to deliver 1.5x the daily returns of the S&P 500 VIX Short-Term Futures Index. This index represents a rolling long position in the first and second-month VIX futures contracts. UVXY is designed for short-term exposure to volatility, and it typically experiences high volatility itself due to being 1.5x leveraged.

Like VXX, UVXY is tied to VIX futures rather than the spot VIX, but it amplifies its returns (and losses) through leverage.

Since UVXY aims to maintain a constant 1.5x leverage, it undergoes daily rebalancing. This daily rebalancing leads to what is known as "volatility decay", where the compounding effect can erode returns in choppy markets.

Below is a graph comparing the prices of VXX and UVXY:


```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for UVXY
uvxy_data = yf.download("UVXY", start="2020-01-01")
vxx_data = yf.download("VXX", start="2020-01-01")

# Plot the decay of UVXY
plt.figure(figsize=(14, 7))
plt.plot(uvxy_data.index, uvxy_data['Close'], label='UVXY', color='purple')
plt.plot(vxx_data.index, vxx_data['Close'], label='VXX', color='blue')
plt.title('UVXY Decay Over Time')
plt.xlabel('Date')
plt.ylabel('UVXY Price')
plt.grid(True)
plt.legend()
plt.show()
```

### Risks and Rewards
While UVXY can provide significant returns in short bursts, it carries substantial risks:

- Decay Over Time: Due to daily rebalancing and the effects of contango, UVXY is prone to significant decay over time, making it unsuitable for long-term holding.
- Volatility Drag: The 1.5x leverage works well during trending markets but amplifies losses during sideways or volatile periods. This drag is increased when volatility is high but directionless.

### Trading strategies involving UVXY
UVXYs high volatility and leverage make it attractive for specific trading strategies. Here are a few examples:

- Day Trading: Given UVXYs extreme volatility and quick movements, its often used by day traders looking to profit from short-term spikes in market fear. However, holding it overnight can be risky due to its decay.

- Hedging: Traders might buy UVXY as a hedge against sudden market drops or to protect long positions during uncertain times. The idea is to offset losses in a broader portfolio with UVXYs sharp gains during a market crash.

- Mean Reversion: Some traders take advantage of UVXYs tendency to spike and then revert back to its downward trend. They may short UVXY after a spike, betting on a quick reversion.

In summary, UVXY can be a useful ETF to trade during periods of volatility. Due to its leverage compared to VXX, it allows for amplified returns, but its risks require careful management.


## Conclusion
VIX can be a userful indicator for times of volatility. VXX and UVXY are powerful tools for trading volatility, but they come with significant risks. While they can provide substantial short-term gains during periods of market stress, their long-term performance is hindered by decay and volatility drag. Understanding how these products work, the risks they carry, and how they fit into various trading strategies is crucial for anyone looking to incorporate them into their portfolio.
